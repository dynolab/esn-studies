```python
import os
import sys
sys.path.append(os.getcwd())

import restools
from comsdk.research import Research

import matplotlib.pyplot as plt
import pickle
import numpy as np
import pandas as pd
```

```python
res = Research.open('RC_MOEHLIS')
```

```python
task_num = 52
data5 = []
time = []
n_files = 100

for num_data in range(n_files):
    with open(os.path.join(res.get_task_path(task_num), str(num_data+1)), 'rb') as a:
        b = pickle.load(a)
    data5.append(b['timeseries'])
    time.append(b['time'])
```

```python
nf = [59, 65]
ui = 0

plt.figure(figsize=(28,10))
for n_f in nf:
    c=0
    for i in range(9):
        plt.subplot(3, 6, i+c*3+1)
        plt.plot(
            np.array(data5[n_f])[:, i], 
            '-', 
            label=str(n_f), 

            );
        plt.xlabel(fr'$L_{"{timeseries}"}$')
        plt.ylabel(fr'$U_{i+1}$')
        #plt.semilogx()
        plt.grid()
        plt.legend();
        
        
        plt.subplot(3, 6, i+3*(c+1)+1)
        uj = i
        plt.plot(
            np.array(data5[n_f])[:, ui], 
            np.array(data5[n_f])[:, uj], 
            '-', 
            label=str(n_f), 

            );

        if (i+1)%3==0:
            c+=1
            
        plt.xlabel(fr'$U_{ui+1}$')
        plt.ylabel(fr'$U_{uj+1}$')
        plt.grid()
        plt.legend();
        
```
